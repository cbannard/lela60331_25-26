```python
import numpy as np
import matplotlib.pyplot as plt
```

**Bandit with stationary rewards**

```python
def n_armed_bandit(k,eps):
    rewards_received=np.zeros((1000, 2000))
    optimal_action_chosen=np.zeros((1000, 2000))
    for run in range(0,2000):
        R_means=np.random.normal(0,1,k)
        Q = [0.0] * k
        N = [0.0] * k
        for i in range(1,1000):
          #print(i)
          if np.random.rand() <= eps:
              A = np.random.randint(0,k)
          else:
              A = Q.index(max(Q))
          R=np.random.normal(R_means[A],1)
          optimal_action_chosen[i,run]=int(R_means.argmax()==A)
          rewards_received[i,run]=R
          N[A] += 1
          Q[A] += ((1/N[A])*(R-Q[A]))
    return(([np.mean(vec) for vec in rewards_received],[np.mean(vec) for vec in optimal_action_chosen]))
```

```python
no_exploration=n_armed_bandit(10,0)
small_exploration=n_armed_bandit(10,0.01)
medium_exploration=n_armed_bandit(10,0.1)
```

```python
plt.plot(no_exploration[0])
plt.xlabel("Trial")
plt.ylabel("Reward")
plt.plot(small_exploration[0])
plt.xlabel("Trial")
plt.ylabel("Reward")
plt.plot(medium_exploration[0])
plt.xlabel("Trial")
plt.ylabel("Reward")
```

```python
plt.plot(no_exploration[1])
plt.xlabel("Trial")
plt.ylabel("Proportion Optimal Chosen")
plt.plot(small_exploration[1])
plt.xlabel("Trial")
plt.ylabel("Proportion Optimal Chosen")
plt.plot(medium_exploration[1])
plt.xlabel("Trial")
plt.ylabel("Proportion Optimal Chosen")
```

**Bandit with non-stationary rewards and constant step size**

```python
def n_armed_bandit(k,eps):
    rewards_received=np.zeros((1000, 2000))
    optimal_action_chosen=np.zeros((1000, 2000))
    for run in range(0,2000):
        R_means=np.zeros(k)
        Q = [0.0] * k
        N = [0.0] * k
        for i in range(1,1000):
          #print(i)
          R_means = np.add(R_means,np.random.normal(0,0.01))
          if np.random.rand() <= eps:
              A = np.random.randint(0,k)
          else:
              A = Q.index(max(Q))
          R=R_means[A]
          optimal_action_chosen[i,run]=int(R_means.argmax()==A)
          rewards_received[i,run]=R
          N[A] += 1
          Q[A] += ((1/N[A])*(R-Q[A]))
    return(([np.mean(vec) for vec in rewards_received],[np.mean(vec) for vec in optimal_action_chosen]))
```

```python
no_exploration=n_armed_bandit(10,0)
small_exploration=n_armed_bandit(10,0.01)
medium_exploration=n_armed_bandit(10,0.1)
```

```python
plt.plot(no_exploration[0])
plt.xlabel("Trial")
plt.ylabel("Reward")
plt.plot(small_exploration[0])
plt.xlabel("Trial")
plt.ylabel("Reward")
plt.plot(medium_exploration[0])
plt.xlabel("Trial")
plt.ylabel("Reward")
```
